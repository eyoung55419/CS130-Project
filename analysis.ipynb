```python
import pandas as pd
import numpy as np
df = pd.read_csv("super_bowl_ads.csv")

df["Final_Score"] = (
    0.45 * df["Data.Viewership.Views"] +
    0.35 * df["Data.Viewership.Likes"] -
    0.20 * df["Data.Viewership.Dislikes"] 
)

best_ads = df.sort_values("Final_Score", ascending=False)[["Data.Title", "Final_Score"]]

print("Top 10 Super Bowl Ads of All Time:")
print(best_ads.head(10))
```

```python
top_views = df.sort_values("Data.Viewership.Views", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top_views["Brand"], top_views["Data.Viewership.Views"])
plt.title("Top 10 Super Bowl Ads by Total Views")
plt.xlabel("Views")
plt.ylabel("Brand")
plt.show()
```

```python
top_likes = df.sort_values("Data.Viewership.Likes", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top_likes["Brand"], top_likes["Data.Viewership.Likes"])
plt.title("Top 10 Super Bowl Ads by Likes")
plt.xlabel("Likes")
plt.ylabel("Brand")
```

```python


top_score = df.sort_values("Final_Score", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.barh(top_score["Brand"], top_score["Final_Score"])
plt.title("Top 10 Super Bowl Ads by Combined Score")
plt.xlabel("Final Score")
plt.ylabel("Brand")
```
